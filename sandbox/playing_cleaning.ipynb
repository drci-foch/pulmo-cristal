```python
import pandas as pd
```

```python
path = r"C:\Users\benysar\Documents\GitHub\pulmo-cristal\donneurs_data_v0.1.0_20250521_103320.csv"
df = pd.read_csv(
    path,
    sep=";",
)
```

```python
# Modifier les valeurs selon les conditions
df["Type de donneur"] = df["Type de donneur"].apply(
    lambda x: "SME"
    if "SME" in x
    else "DDAC M3"
    if "DDAC M3" in x
    else "DDME"
    if "DDME" in x
    else x
)
```

```python
df[["Groupe", "Rhésus"]] = df["Groupe sanguin"].str.extract(r"([ABOabo]+)\s*([+-])")

# Mettre la colonne Groupe en majuscules
df["Groupe"] = df["Groupe"].str.upper()

# Réorganiser les colonnes
# Insérer les nouvelles colonnes après 'Groupe sanguin'
cols = list(df.columns)
index = cols.index("Groupe sanguin")
new_order = cols[: index + 1] + ["Groupe", "Rhésus"] + cols[index + 1 :]
df = df[new_order]
df = df.iloc[:, :-3]
```

```python
import numpy as np

df["Poids (kg)"] = np.where(
    df["Poids (kg)"] % 1 >= 0.5, np.ceil(df["Poids (kg)"]), np.floor(df["Poids (kg)"])
).astype(int)

df
```

```python
# Modifier les valeurs selon les conditions
df["Arrêt cardiaque récupéré"] = df["Arrêt cardiaque récupéré"].apply(
    lambda x: "Non" if "Non" in x else "Oui" if "Oui" in x else x
)
df["Arrêt cardiaque récupéré"]
```

```python
# Modifier les valeurs selon les conditions
df["Traumatisme broncho-pulmonaire actuel"] = df[
    "Traumatisme broncho-pulmonaire actuel"
].apply(
    lambda x: "Non"
    if "Non" in x
    else "Oui"
    if "Oui" in x
    else "NA"
    if "Non renseigné"
    else "NA"
)
df["Traumatisme broncho-pulmonaire actuel"]
```

```python
df["Lésion pleurale traumatique actuelle"]
```

```python
df["Lésion pleurale traumatique actuelle"] = df[
    "Lésion pleurale traumatique actuelle"
].apply(
    lambda x: "Non"
    if "Non" in x
    else "Oui"
    if "Oui" in x
    else "NA"
    if "Non renseigné"
    else "NA"
)
```

```python
colonnes = [
    "PaCO2 (mmHg)",
    "PaO2 (mmHg)",
    "CO3H- (mmol/l)",
    "SaO2 (%)",
    "Fraction d'éjection",
]
df[colonnes] = df[colonnes].round(0)
```

```python
df["Fraction d'éjection"] = pd.to_numeric(df["Fraction d'éjection"], errors='coerce')
```

```python
df.to_excel("cristal_structuration_cleaned.xlsx")
```
